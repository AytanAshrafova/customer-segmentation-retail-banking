```python
# 2.1 Boş dəyərlərin yoxlanması
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome', 'y']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print(df.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# 5.1 Optimal klaster sayını tapmaq üçün Elbow Method istifadə edək
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE (Sum of squared errors)')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)
print(df['Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df.drop('Cluster', axis=1))

plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_result[:,0], y=pca_result[:,1], hue=df['Cluster'], palette='Set2')
plt.title('Customer Segments Visualized by PCA')
plt.show()
```

```python

print(df['housing'].value_counts())



df_filtered = df[df['housing'] != 2]

for cluster in sorted(df_filtered['Cluster'].unique()):
    subset = df_filtered[df_filtered['Cluster'] == cluster]
    print(f"Cluster {cluster}:")
    print(f"Number of customers: {len(subset)}")
    print(f"Average age: {subset['age'].mean():.2f}")
    print(f"Proportion with housing loan: {subset['housing'].mean() * 100:.2f}%")
    print(f"Proportion subscribed to term deposit (y = yes): {subset['y'].mean() * 100:.2f}%")
    print("------")
```

```python
print("Business insights per cluster:\n")

for cluster in sorted(df_filtered['Cluster'].unique()):
    subset = df_filtered[df_filtered['Cluster'] == cluster]
    n_customers = len(subset)
    avg_age = subset['age'].mean()
    housing_loan_rate = subset['housing'].mean() * 100
    deposit_sub_rate = subset['y'].mean() * 100

    print(f"Cluster {cluster}:")
    print(f"- Number of customers: {n_customers}")
    print(f"- Average age: {avg_age:.2f}")
    print(f"- Proportion with housing loan: {housing_loan_rate:.2f}%")
    print(f"- Proportion subscribed to term deposit (y=yes): {deposit_sub_rate:.2f}%")

    if deposit_sub_rate > 20:
        print("- This cluster shows high subscription rates, good for targeted marketing campaigns.")
    elif deposit_sub_rate < 5:
        print("- Low subscription rate; might need more engagement or tailored product offers.")
    else:
        print("- Moderate subscription; potential for growth with personalized offers.")

    print("------\n")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X, kmeans.labels_))

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')

plt.subplot(1,2,2)
plt.plot(K, silhouette, 'rx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')

plt.show()
```
